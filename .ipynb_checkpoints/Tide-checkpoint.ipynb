```python
import pandas as pd
import numpy as np
```

```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen
```

```python
url = "https://tidesandcurrents.noaa.gov/cdata/StationList?type=Current+Data&filter=active"
urls = []
```

```python
for var in range(1,2):
    url = ["https://tidesandcurrents.noaa.gov/cdata/StationList?type=Current+Data&filter=active"]
    urls.append(url[0])
```

```python
urls
```

```python
def connect (ur):
    uClient = urlopen(ur)
    page_html = uClient.read()
    
    uClient.close()
    
    page_soup = soup(page_html, "html.parser")
    
    container = page_soup.find_all("table", {"class": "table table-condensed table-striped table-bordered"})
    return container
```

```python
con = connect(urls[0])
print(soup.prettify(con[0]))
```

```python
def dataframe (containers):
    data = {"Station_Name":[],"Station_ref_code":[]}
    for i in range(0,81):
        con = connect(urls[0])
        for container in con:
            station = container.findAll('a', {"target":"_blank"})
            st_a = station[i].text
            data["Station_Name"].append(st_a)

            info = container.findAll('a',{"rel":"nofollow"})
            ref_code = info[i].text
            data["Station_ref_code"].append(ref_code)
        
        print("done",i)

    return data

```

```python
dataset = dataframe(urls)
```

```python
df = pd.DataFrame(dataset)
df.drop_duplicates(inplace=True,ignore_index=True)
df
```

```python
df.to_csv("Stationdetails.csv")
```
