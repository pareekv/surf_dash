```python
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen
```

```python
url = "https://tidesandcurrents.noaa.gov/cdata/StationList?type=Current+Data&filter=active"
urls = []
```

```python
for var in range(1,2):
    url = ["https://tidesandcurrents.noaa.gov/cdata/StationList?type=Current+Data&filter=active"]
    urls.append(url[0])
```

```python
urls
```

```python
def connect (ur):
    uClient = urlopen(ur)
    page_html = uClient.read()
    
    uClient.close()
    
    page_soup = soup(page_html, "html.parser")
    
    container = page_soup.find_all("table", {"class": "table table-condensed table-striped table-bordered"})
    return container
```

```python
con = connect(urls[0])
print(soup.prettify(con[0]))
```
